````python
import pandas as pd

# Load the main remote data file into a DataFrame
df = pd.read_csv('DRYAD_Data_Johansson_Updated2_1.csv')

df['Time of Detection'] = pd.to_datetime(df['Time of Detection']).dt.time

# Create dataset_2000_2359 CSV
# Filter rows where the time is between 20:00 (8:00 PM) and 23:59 (11:59 PM)
filtered_data1 = df[(df['Time of Detection'] >= pd.to_datetime('20:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('23:59:59').time())]
# Group by species and date, count, unstack and Fill binary
filtered_data1 = filtered_data1.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_2000_2359.csv fileS
filtered_data1.to_csv('dataset_2000_2359.csv', index=True)


# Create dataset_0000_0459 CSV
# Filter rows where the time is between 00:00 (00:00 AM) and 04:59 (04:59 AM)
filtered_data2 = df[(df['Time of Detection'] >= pd.to_datetime('00:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('04:59:59').time())]
# Group by species and date, count, unstack and Fill binary
filtered_data2 = filtered_data2.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_0000_0459.csv fileS
filtered_data2.to_csv('dataset_0000_0459.csv', index=True)


# Create dataset_2000_2359_urban CSV
# Filter rows where the time is between 20:00 (8:00 PM) and 23:59 (11:59 PM) and Forest Cover within 1.5km of camera (km^2) less than 0.3
filtered_data3 = df[(df['Time of Detection'] >= pd.to_datetime('20:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('23:59:59').time()) & (df['Forest Cover within 1.5km of camera (km^2)'] < 0.3)]
# Group by species and date, count, unstack and Fill binary
filtered_data3 = filtered_data3.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_2000_2359_urban.csv fileS
filtered_data3.to_csv('dataset_2000_2359_urban.csv', index=True)


# Create dataset_2000_2359_rural CSV
# Filter rows where the time is between 20:00 (8:00 PM) and 23:59 (11:59 PM) and Forest Cover within 1.5km of camera (km^2) greater than or equal to 0.3
filtered_data4 = df[(df['Time of Detection'] >= pd.to_datetime('20:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('23:59:59').time()) & (df['Forest Cover within 1.5km of camera (km^2)'] >= 0.3)]
# Group by species and date, count, unstack and Fill binary
filtered_data4 = filtered_data4.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_2000_2359_rural.csv fileS
filtered_data4.to_csv('dataset_2000_2359_rural.csv', index=True)


# Create dataset_0000_0459_urban CSV
# Filter rows where the time is between 00:00 (00:00 AM) and 04:59 (04:59 AM) and and Forest Cover within 1.5km of camera (km^2) less than 0.3
filtered_data5 = df[(df['Time of Detection'] >= pd.to_datetime('00:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('04:59:59').time()) & (df['Forest Cover within 1.5km of camera (km^2)'] < 0.3)]
# Group by species and date, count, unstack and Fill binary
filtered_data5 = filtered_data5.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_0000_0459_urban.csv fileS
filtered_data5.to_csv('dataset_0000_0459_urban.csv', index=True)


# Create dataset_0000_0459_rural CSV
# Filter rows where the time is between 00:00 (00:00 AM) and 04:59 (04:59 AM) and and Forest Cover within 1.5km of camera (km^2) greater than or equal to 0.3
filtered_data6 = df[(df['Time of Detection'] >= pd.to_datetime('00:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('04:59:59').time()) & (df['Forest Cover within 1.5km of camera (km^2)'] >= 0.3)]
# Group by species and date, count, unstack and Fill binary
filtered_data6 = filtered_data6.groupby(['Species Detected', 'Date of Detection']).count().loc[:, 'Yard'].unstack().fillna(False).map(lambda x: not x).map(lambda x: not x)
# Save the result to dataset_0000_0459_rural.csv fileS
filtered_data6.to_csv('dataset_0000_0459_rural.csv', index=True)
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("DRYAD_Data_Johansson_Updated2_1.csv")
df['Time of Detection'] = pd.to_datetime(df['Time of Detection']).dt.time
df['Hour of Detection'] = df['Time of Detection'].apply(lambda t: t.hour)

# What are the top 5 species detected over the entire dataset?
# Group by Species Detected and sum up the counts
grouped_df = df.groupby('Species Detected')['Number of the Species Detected in the Photo'].sum().reset_index()
# Sort by count in descending order and take the top 5
top_5_species = grouped_df.sort_values(by='Number of the Species Detected in the Photo', ascending=False).head(5)
print("Top 5 Species Detected are:")
print(top_5_species['Species Detected'])


# Which 5 species over all data are most frequent at midnight (12am)?
# Filter rows where the Hour of Detection is 0 which means midnight
midnight_df = df[df['Hour of Detection'] == 0]
# Group by species and sum up the counts
midnight_df = midnight_df.groupby('Species Detected')['Number of the Species Detected in the Photo'].sum().reset_index()
# Sort by count in descending order and take the top 5
midnight_df_top_5 = midnight_df.sort_values(by='Number of the Species Detected in the Photo', ascending=False).head(5)
print("Top 5 Species Detected at Midnight are:")
print(midnight_df_top_5['Species Detected'])


# Use the library sns.heatmap() to generate a heatmap with Species Detected on the y-axis and Hour of Detection on the x-axis.
# Create a DataFrame that shows the frequency of each species per hour
heatmap_data = df.groupby(['Species Detected', 'Hour of Detection']).size().reset_index(name='Frequency')
# Pivot this DataFrame so that we have Species on the y-axis and Hour on the x-axis
heatmap_data_pivot = heatmap_data.pivot(index="Species Detected", columns="Hour of Detection", values="Frequency").fillna(0)

# Generate the heatmap
plt.figure(figsize=(27, 18))
sns.heatmap(heatmap_data_pivot, annot=True, cmap="crest")
plt.title("Species Detected by Hour of Day")
plt.show()


# Compare the result of your prior answer with running heatmap() with the robust=True parameter. Which of the two output graphs looks more interpretable. Why?
# Generate the heatmap
plt.figure(figsize=(27, 18))
sns.heatmap(heatmap_data_pivot, annot=True, cmap="crest", robust=True)
plt.title("Species Detected by Hour of Day(With robust True)")
plt.show()


# Compute the mode of all species in the dataset.
# Find the mode of the 'Species Detected'
mode_species = df['Species Detected'].mode()[0]
print(f"The mode of the 'Species Detected' is: {mode_species}")


#Compute the probabilities of all species, but make two columns: (1) with those in the group with forest density < 0.3 and (2) those with forest density  0.3. Label them urban and rural.
# Create a new column indicating 'urban' or 'rural' based on forest density
df['DensityType'] = df['Forest Cover within 1.5km of camera (km^2)'].apply(lambda x: 'urban' if x < 0.3 else 'rural')

# Compute probabilities for each group
total_counts = df.groupby('DensityType')['Number of the Species Detected in the Photo'].sum()
grouped_df = df.groupby(['Species Detected', 'DensityType'])['Number of the Species Detected in the Photo'].sum().reset_index()
grouped_df['Probability'] = grouped_df.apply(lambda row: row['Number of the Species Detected in the Photo'] / total_counts[row['DensityType']], axis=1)

# Pivot the DataFrame to get 'urban' and 'rural' as columns
pivot_df = grouped_df.pivot(index='Species Detected', columns='DensityType', values='Probability').fillna(0)
print(pivot_df)
````

This approach places emphasis on the predominant range of values in the dataset. As a result, the heatmap becomes more straightforward to comprehend, particularly when the dataset includes a small number of extreme outliers that can distort the color representation.

If our dataset incorporates extreme outliers that are less relevant for our analytical objectives, opting for robust=True in the heatmap configuration may lead to a more intelligible visualization. This is because it refines the color scale to align with the more frequently occurring data values.

Conversely, if outliers are integral and meaningful to your specific case, the heatmap without the robust=True setting may be the more suitable choice. In this scenario, the color scale considers these outlier values when establishing its range, ensuring that they are appropriately represented in the heatmap.

In summary, the robust=True parameter in sns.heatmap() allows us to fine-tune the heatmap's color scale by accounting for or ignoring extreme data values based on the specific requirements of your analysis.

````python
import pandas as pd
from mlxtend.frequent_patterns import apriori

# Read the dataset
df = pd.read_csv("DRYAD_Data_Johansson_Updated2_1.csv")
df['Time of Detection'] = pd.to_datetime(df['Time of Detection']).dt.time
df['Hour of Detection'] = df['Time of Detection'].apply(lambda t: t.hour)

def create_binary_matrix(df):
    grouped_df = df.groupby(['Species Detected', 'Date of Detection']).size().reset_index(name='Count')
    binary_df = pd.pivot_table(grouped_df, values='Count', index='Date of Detection', columns='Species Detected', fill_value=False)
    binary_df[binary_df > 1] = True
    return binary_df

# Which early evening species are frequent together in groups of 2 or more regardless of rural or urban? (Use a support of 0.50 when answering).
early_evening_groups_2_df = df[(df['Time of Detection'] >= pd.to_datetime('21:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('23:59:59').time()) & (df['Number of the Species Detected in the Photo'] >= 2)]
early_evening_groups_2_df = early_evening_groups_2_df.groupby(['Species Detected', 'Date of Detection']).size().reset_index(name='Count')
# Convert to a binary matrix
early_evening_groups_2_df = pd.pivot_table(early_evening_groups_2_df, values='Count', index='Date of Detection', columns='Species Detected', fill_value=False)
early_evening_groups_2_df[early_evening_groups_2_df >= 1] = True
early_evening_groups_2_df = apriori(early_evening_groups_2_df, min_support=0.50, use_colnames=True)
print("Early evening species which are frequent together in groups of 2 or more regardless of rural or urban? are: ")
print(early_evening_groups_2_df['itemsets'])


print("\n")
# Show the early evening species frequent patterns for both urban and rural. Compare and contrast them. Do you see any surprises or are these what you expect?
early_evening_df = df[(df['Time of Detection'] >= pd.to_datetime('21:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('23:59:59').time())]
urban_early_evening_df = early_evening_df[early_evening_df['Forest Cover within 1.5km of camera (km^2)'] < 0.3]
rural_early_evening_df = early_evening_df[early_evening_df['Forest Cover within 1.5km of camera (km^2)'] >= 0.3]
urban_early_evening_binary = create_binary_matrix(urban_early_evening_df)
rural_early_evening_binary = create_binary_matrix(rural_early_evening_df)
urban_early_evening_frequent_itemsets = apriori(urban_early_evening_binary, min_support=0.50, use_colnames=True)
rural_early_evening_frequent_itemsets = apriori(rural_early_evening_binary, min_support=0.50, use_colnames=True)
print("Early evening species frequent patterns for urban")
print(urban_early_evening_frequent_itemsets)
print("Early evening species frequent patterns for rural")
print(rural_early_evening_frequent_itemsets)


print("\n")
# Show the late evening species frequent patterns for both urban and rural. Compare and contrast them. Do you see any surprises or are these what you expect?
late_evening_df = df[(df['Time of Detection'] >= pd.to_datetime('00:00:00').time()) & (df['Time of Detection'] <= pd.to_datetime('4:00:00').time())]
urban_late_evening_df = late_evening_df[late_evening_df['Forest Cover within 1.5km of camera (km^2)'] < 0.3]
rural_late_evening_df = late_evening_df[late_evening_df['Forest Cover within 1.5km of camera (km^2)'] >= 0.3]
urban_late_evening_binary = create_binary_matrix(urban_late_evening_df)
rural_late_evening_binary = create_binary_matrix(rural_late_evening_df)
urban_late_evening_frequent_itemsets = apriori(urban_late_evening_binary, min_support=0.50, use_colnames=True)
rural_late_evening_frequent_itemsets = apriori(rural_late_evening_binary, min_support=0.50, use_colnames=True)
print("Late evening species frequent patterns for urban")
print(urban_late_evening_frequent_itemsets)
print("Late evening species frequent patterns for rural")
print(rural_late_evening_frequent_itemsets)


print("\n")
# (Reflection)
# Fill in the table with the most frequent itemset (with length >1) of highest support for each:
# Apply Apriori algorithm
def get_high_support_itemsets(df):
    frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
    return frequent_itemsets.sort_values('support', ascending=False)


# Get the highest support value with more than one item
def get_highest_support(frequent_itemsets):
    frequent_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]
    return frequent_itemsets.head(1)


early_evening_urban_frequent = get_high_support_itemsets(urban_early_evening_binary)
late_evening_urban_frequent = get_high_support_itemsets(urban_late_evening_binary)
early_evening_rural_frequent = get_high_support_itemsets(rural_early_evening_binary)
late_evening_rural_frequent = get_high_support_itemsets(rural_late_evening_binary)
urban_early_evening_highest_support = get_highest_support(early_evening_urban_frequent)
urban_late_evening_highest_support = get_highest_support(late_evening_urban_frequent)
rural_early_evening_highest_support = get_highest_support(early_evening_rural_frequent)
rural_late_evening_highest_support = get_highest_support(late_evening_rural_frequent)
# Create the table with headers
table = pd.DataFrame({
    '': ['Urban', 'Rural'],
    'early evening (9p-1159p)': [urban_early_evening_highest_support['itemsets'].to_string(index=False) + '; sup = ' + urban_early_evening_highest_support['support'].to_string(index=False),
                                 rural_early_evening_highest_support['itemsets'].to_string(index=False) + '; sup = ' + rural_early_evening_highest_support['support'].to_string(index=False)],
    'late evening (12a-4a)': [urban_late_evening_highest_support['itemsets'].to_string(index=False) + '; sup = ' + urban_late_evening_highest_support['support'].to_string(index=False),
              rural_late_evening_highest_support['itemsets'].to_string(index=False) + '; sup = ' + rural_late_evening_highest_support['support'].to_string(index=False)]
})
print(table.to_string(index=False))
````
